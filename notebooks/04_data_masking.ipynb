```python
import geopandas as gpd
```

```python
warsaw_tiles = gpd.read_file('../outputs/warsaw_coverage.geojson')
warsaw_tiles.crs = "EPSG:4326"
```

```python
orange_gdf = gpd.read_file('../outputs/operators_5g_coverage.geojson')
orange_gdf.crs = "EPSG:4326"
```

```python
# index_right column creates conflict
warsaw_tiles = warsaw_tiles.drop(columns=['index_right'])
```

```python
operators_covered_tiles = gpd.sjoin(warsaw_tiles, orange_gdf, how="inner", predicate='intersects')
```

```python
display(operators_covered_tiles.head())
```

```python
operators_covered_tiles.to_file('../outputs/operators_5g_coverage_warsaw.geojson', driver='GeoJSON')
```

```python
orange_5g_coverage_warsaw = operators_covered_tiles[operators_covered_tiles['operator'] == 'Orange']
```

```python
display(orange_5g_coverage_warsaw.head())
```

```python
orange_5g_coverage_warsaw.to_file('../outputs/orange_5g_coverage_warsaw.geojson', driver='GeoJSON')
```
